```python
import requests
import pandas as pd
import json
import time
```

### Get ID list

```python
url = 'https://ratings-api.ea.com/v2/entities/fifa-23-womens-rating?sort=ranking%3AASC&limit=1200'
```

```python
r = requests.get(url)
```

```python
json_data = json.loads(r.text)
```

```python
data_docs = json_data['docs']
```

```python
data_docs[0]
```

```python
length = len(data_docs)
```

```python
id_list = []
```

```python
for i in range(length):
    id_list.append(data_docs[i]['primaryKey'])
```

```python
len(id_list)
```

### Get Dataframe

```python
df_rating = pd.DataFrame()
```

```python
def get_rating(df, x):
    url = 'https://ratings-api.ea.com/v2/entities/fifa-23-womens-rating?filter=primaryKey%3A'+ x +'&sort=ranking%3AASC&limit=40'
    r = requests.get(url)
    text = r.text
    json_data = json.loads(text)
    data = json_data['docs']
    df_temp = pd.DataFrame.from_dict(data)
    
    df = pd.concat([df, df_temp], ignore_index = True)
    return df
```

```python
for i in id_list:
    df_rating = get_rating(df_rating, str(i))
    time.sleep(0.5)
```

```python
df_rating.to_csv('FIFA23_rating/player_rating.csv', index = False)
```
